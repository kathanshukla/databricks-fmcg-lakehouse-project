```python
from pyspark.sql import functions as F
from pyspark.sql.functions import concat as concate

start_date = '2024-01-01'
end_date = '2025-12-01'
```

```python
date_df = spark.sql(f"SELECT explode(sequence(to_date('{start_date}'), to_date('{end_date}'), interval 1 day)) AS date")
display(date_df)
```

```python
df = (
    spark.sql(f"""
        SELECT explode(
            sequence(
                to_date('{start_date}'),
                to_date('{end_date}'),
                interval 1 month
            )
        ) AS month_start_date
    """)
)

display(df)
```

```python
date_table = df.withColumn("date_key", F.date_format("month_start_date", "yyyyMM").cast("int"))\
    .withColumn("Year", F.year(F.col('month_start_date'))) \
    .withColumn("Month_name", F.date_format("month_start_date", "MMMM")) \
    .withColumn("Month_short_name", F.date_format("month_start_date", "MMM")) \
    .withColumn("Quarter", F.concat(F.lit("Q"),F.quarter(F.col('month_start_date')))) \
    .withColumn("year_quarter",F.concat(F.year(F.col('month_start_date')),F.lit("-Q"),F.quarter(F.col('month_start_date')))) 
# display(date_table)
```

```python
display(date_table)
```

```python
date_table.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("fmcg.gold.dim_date")
```
